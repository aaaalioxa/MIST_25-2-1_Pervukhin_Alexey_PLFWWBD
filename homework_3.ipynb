## Задача 1

```python
def count_word_frequency(text):
    text = text.lower()
    
    for znak in [".", ",", "!", "?", ":", ";", "-", "'"]:
        text = text.replace(znak, "")
    
    words = text.split()

    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    
    return frequency
```

```python
text = "Я люблю МИСИС. Просто я обожаю Мисис"
frequency_dict = count_word_frequency(text)
print(frequency_dict)
```

```python
def calculate_price(price, discount=0, tax_rate=0.2):
    return price * (1 - discount / 100) * (1 + tax_rate)
```

## Задача 3

```python
import datetime

def analyze_log(filepath):

    result = {
        'counts': {'INFO': 0, 'WARNING': 0, 'ERROR': 0},
        'time_range': (None, None)
    }

    min_date = None
    max_date = None

    try:
        with open(filepath, "r") as f:
            lines_list = f.readlines()
    except FileNotFoundError:
        print(f"Ошибка: Файл по пути '{filepath}' не найден.")

    for line in lines_list:
        try:
            parts = line.split(",")

            date = parts[0]
            type = parts[1]

            date = datetime.datetime.strptime(date, "%Y-%m-%d %H:%M:%S")

            if type in result["counts"]:
                result["counts"][type] += 1

            if min_date is None or date < min_date:
                min_date = date

            if max_date is None or date > max_date:
                max_date = date

        except (IndexError, ValueError):
            continue

    if min_date and max_date:
            result["time_range"] = (
                min_date.strftime("%Y-%m-%d %H:%M:%S"),
                max_date.strftime("%Y-%m-%d %H:%M:%S")
            )

    return result
```

```python
analyze_log("log.txt")
```
