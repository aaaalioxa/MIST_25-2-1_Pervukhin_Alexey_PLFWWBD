## Задача 1

```python
n = int(input())

for i in range(1, n+1):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz", end = ' ')
    elif i % 3 == 0:
        print("Fizz", end = ' ')
    elif i % 5 == 0:
        print("Buzz", end = ' ')
    else:
        print(i, end = ' ')
        
```

## Задача 2

```python
shopping_list = []

while True:

    print('''
          Меню:
          1. Добавить
          2. Показать
          3. Выйти''')
    
    choice = int(input("Введите номер опции:"))

    if choice == 1:
        shopping_list.append(input("Введите название продукта:"))
    elif choice == 2:
        if len(shopping_list) == 0:
            print("Список пуст")
        else:
            for num, val in enumerate(shopping_list):
                print(f"{num + 1}. {val}")
    elif choice == 3:
        print("Программа завершена")
        break
    else:
        print("Неизвестная команда, попробуйте снова")
    
```
